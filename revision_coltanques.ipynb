```python
data_folders= [f for f in os.listdir(br_folders) if '.rar' not in f] 
data_folders

raw_data_paths= []
data_names= []
for i in range(len(data_folders)):
    temp_data_paths= os.path.join(br_folders, data_folders[i])
    raw_data_paths.append(temp_data_paths)
    data_names.append(data_folders[i])
raw_data_paths
```

```python
##### 
lat_var='Latitud' 
lon_var= 'Longitud'
placa_var= "Nombre del Vehículo"
velocidad_var= "Velocidad [km/h]"

x= 0  ### cuanta las iteraciones

for j in range(len(raw_data_paths)):
    if raw_data_paths[j].endswith("csv"): 
       gps_temp= pd.read_csv(raw_data_paths[j], delimiter= ",") 
    elif  raw_data_paths[j].endswith("xlsx"):
       gps_temp= pd.read_excel(raw_data_paths[j])  

    ### registrar el nombre de las variables antes de introducir otras
    colums= ""
    for k in range(len(gps_temp.columns)):
      colums= colums + gps_temp.columns[k] + "," 
    ### velocidad
    velocity= gps_temp.groupby(placa_var)[[velocidad_var]].agg(['mean', 'min', 'max'])
    velocity.reset_index(inplace=True)
    colnames= [ f[0] +"_" + f[1]  for f in  velocity.columns]
    velocity.columns= colnames
    tem_varname= placa_var +"_"
    velocity.rename(columns={tem_varname: "PLACA"}, inplace= True)
    velocity['Rango Velocidad']= ""
    for t in range(len(velocity)):
        a=velocity.loc[t,  velocidad_var+'_min']
        b=velocity.loc[t, velocidad_var+'_max']
        velocity.loc[t, 'Rango Velocidad']= "("+ str(a) + "-" + str(b) + ")"
    
    velocity= velocity[['PLACA', velocidad_var+ '_mean', "Rango Velocidad" ]]

    gps_temp['velocidad_valida']= 1
    gps_temp.loc[gps_temp[velocidad_var]> 125, 'velocidad_valida']= 0
    gps_temp.loc[gps_temp[velocidad_var]< 0, 'velocidad_valida']= 0    ### velocidades menores a cero
    for var in [lat_var , lon_var]:
            varname= var + "_valida"
            gps_temp[varname]= 1
            if var== lon_var:
                    gps_temp.loc[abs(gps_temp[var])> 180, 'velocidad_valida']= 0
            elif var== lat_var:
                    gps_temp.loc[abs(gps_temp[var])> 90, 'velocidad_valida']= 0

    gps_temp['obs_valida']= 1

    lat_valida=  lat_var + "_valida"
    lon_valida=  lon_var + "_valida"

    gps_temp.loc[(gps_temp['velocidad_valida']== 0) | (gps_temp[lat_valida]== 0) | (gps_temp[lon_valida]== 0), 'obs_valida']= 0

    obs_validas= gps_temp.groupby(placa_var)[['obs_valida']].agg(['mean', "sum"])
    obs_validas.reset_index(inplace=True)
    obs_validas.columns= ["PLACA", "obs_validas (%)", "obs_validas total"]

    regis=gps_temp.groupby(placa_var)[[placa_var]].agg('count')
    regis.columns= ["CANT.REGISTROS"]
    regis.reset_index(inplace=True)
      
    ### latitud longitud
    t=0
    for var in [lat_var , lon_var]:
          tem_data=  gps_temp.groupby(placa_var)[[var]].agg(['min', 'max'])
          tem_data.columns= tem_data.columns.droplevel()
          tem_data.reset_index(inplace=True)
          col_min= var + '_min'
          col_max= var + '_max'
          tem_data.columns= [placa_var,col_min , col_max]
          if t==0:
            lat_lon=tem_data.copy()
          else: 
                lat_lon= pd.merge(lat_lon, tem_data, on= placa_var) 
          t= t+1
    lat_lon['Rango Longitud']= ""
    lat_lon['Rango Latitud']= ""
    for p in range(len(lat_lon)):
          for var in [lat_var , lon_var]:
                col_min= var + '_min'
                col_max= var + '_max'
                a=lat_lon.loc[p, col_min]
                b=lat_lon.loc[p, col_max]
                if var== lon_var:
                      lat_lon.loc[p, 'Rango Longitud']= "("+ str(a) + "," + str(b) + ")"
                elif var== lat_var:
                      lat_lon.loc[p, 'Rango Latitud']= "("+ str(a) + "," + str(b) + ")"

    tem_revision_data= pd.merge(lat_lon[[placa_var, "Rango Longitud", "Rango Latitud" ]], regis, on= placa_var )
    ### Fecha
    today = date.today()
    tem_revision_data["Fecha Revisión"]= today
    ### id
    a= len(tem_revision_data)+ 1
    tem_revision_data["ID Archivo-Placa"]= range(1, a)
    ### nombre archivo-pestaña
    tem_revision_data["Nombre Archivo-Pestaña"]=  data_names[j]
    # PLACA
    tem_revision_data = tem_revision_data.rename(columns={placa_var: 'PLACA'})

    tem_revision_data=pd.merge(tem_revision_data, velocity, on= "PLACA", how= "outer")
    tem_revision_data=pd.merge(tem_revision_data, obs_validas, on= "PLACA", how= "outer")
    #### Nombre de Variables
    tem_revision_data["Nombre Variables"]=  colums
    
    if x==0: 
        revision_data=  tem_revision_data.copy()
    else :
        revision_data= revision_data.append(tem_revision_data)   
    x=x+1    
```

```python
first_vars = ['ID Archivo-Placa', 'PLACA', 'Nombre Archivo-Pestaña', 'Fecha Revisión']   ## variables que queremos de primeras
varlist =[w for w in revision_data.columns if w not in first_vars]
order_var= first_vars+varlist
order_var
```

```python
#### Agregados
agregado= pd.DataFrame( columns= [ 'TOTAL ARCHIVOS/BD',  'ARCHIVOS/BD CORRECTOS', 'TOTAL VEHÍCULOS' , 'TOTAL VEHÍCULOS CORRECTOS',	 'TOTAL REGISTROS',	'TOTAL REGISTROS CORRECTOS', 'PORCENTAGE REGISTROS CORRECTOS' ])

### total archivos
agregado.loc[1, 'TOTAL ARCHIVOS/BD' ]= len(raw_data_paths)

# total vehículos

a= len(revision_data.groupby("PLACA")[["PLACA"]].agg("count"))
agregado.loc[1, 'TOTAL VEHÍCULOS' ]= a

# archivos correctos
agregado.loc[1, 'ARCHIVOS/BD CORRECTOS' ]= len(raw_data_paths)

### perdimos vehículos con observaciones correctas= 0
dummy_df= revision_data.groupby("PLACA")[["obs_validas total"]].agg("sum")
dummy_df.reset_index(inplace=True)
dummy_df["dummy_var"]= 1
dummy_df.loc[dummy_df["obs_validas total"]==0 , "dummy_var"]= 0

a= dummy_df["dummy_var"].sum()

# Total vehículos correctos
agregado.loc[1, 'TOTAL VEHÍCULOS CORRECTOS' ]= a

## total registros
a= revision_data['CANT.REGISTROS'].sum()
agregado.loc[1, 'TOTAL REGISTROS' ]= a

### total registrs válidos

a= revision_data['obs_validas total'].sum()
agregado.loc[1, 'TOTAL REGISTROS CORRECTOS' ]= a

agregado.loc[1, 'PORCENTAGE REGISTROS CORRECTOS' ]= agregado.loc[1, 'TOTAL REGISTROS CORRECTOS' ]/agregado.loc[1, 'TOTAL REGISTROS' ]
```
